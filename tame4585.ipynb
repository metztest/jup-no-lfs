Welcome to RSA!  In this project, I will walk you through using RSA encryption to encrypt and decrypt simple messages.  

In order to make RSA work smoothly and quickly, we will use some mathematical tricks that can save years of processing power. ### 2.1
#### Basic tool set

These are functions that you'll need to pre-process the messages before the messages are encoded and decoded by the RSA algorithm. That is the reason we will be defining them first.



````python
def Convert_Text(_string):
    """
    Define this function such that it takes in a simple 
    string such as "hello" and outputs the corresponding
    standard list of integers (ascii) for each letter in the word hello.
    For example:
    _string = hello
    integer_list = [104, 101, 108, 108, 111]
    """
    integer_list = []
    for i in range(len(_string)):
        cur_num = ord(_string[i])
        integer_list.append(cur_num)
    return integer_list
````

````python
def Convert_Num(_list):
    """
    Do the opposite of what you did in the Convert_Text
    function defined above.
    
    Define this function such that it takes in a list of integers
    and outputs the corresponding string (ascii).
    
    For example:
    _list = [104, 101, 108, 108, 111]
    _string = hello
    """
    _string = ''
    for i in _list:
        _string += chr(i)
    return _string
````

````python
def Convert_Binary_String(_int):
    """
    Here, you need to define a function that converts an integer to
    a string of its binary expansion.
    
    For example:
    _int = 345
    bits = 101011001
    """
    if _int == 0:
        return "0"
    else:
        binary_list = []
        while(_int > 0):
            # grab least significant digit and insert in front of our bits variable
            # starting from 1st place, 2s place, 4s place, etc. until the number can't be divided 
            # by 2 anymore
            digit = _int % 2 
            binary_list.insert(0, str(digit))
            _int = _int // 2
    bits = ''.join(binary_list)
    return bits
````

### 2.2 
#### First tool set.



````python
def FME(b, n, m):
    """
    1. Using the fast modular exponentiation algorithm,
    the below function should return b**n mod m.
    As described on page 254. (however, you may input the exponent and 
    then convert it to a string - the book verison imports the binary expansion)  
    2. You should use the function defined above
    Convert_Binary_String()
    3. Use this string to test which components are used in the calculation.
    4. Yes, there are many other ways to do this. 

***UPDATE*** You may use the function you developed in your problem set - but be sure it is your own work, commented, etc.. and change inputs as needed.
    """  
    
    # We are trying to calculate b^n mod m

    result = 1
    square = b
    while (n > 0):
        k = n % 2  # least significant digit
        n = n // 2
        if k == 1:
            result = (result * square) % m
        square = (square * square) % m
    return result
````

````python
def Euclidean_Alg(a, b):
    """
    1. Calculate the Greatest Common Divisor of a and b.
    
    2. Unless b==0, the result will have the same sign as b (so that when
    b is divided by it, the result comes out positive).
    The function must return a single integer 'x' which is
    the gcd of a and b.
    """
    
    # gcd of a number and 0 is that number
    if b == 0:
        return a 
    if a == 0:
        return b

    x = a
    y = b

    while y != 0: # keep dividing until the remainder is 0 and we're done
        remainder = x % y
        x = y
        y = remainder

    return x
````

### 2.3
#### Second tool set

Here we will implement the meat of the RSA cryptosystem. The functions below will generate the public and private key pairs which will then be used to create a ciphertext using the public key and then decode the same using the pirvate key.



````python
def Find_Public_Key_e(p, q):
    """
    Implement this function such that
    it takes 2 primes p and q.
    
    Use the gcd function that you have 
    defined before.
    
    The function should return 2 elements as follows:
    public key: n
    public key: e
    
    
    HINT: this function will run a loop to find e such 
    that e is relatively prime to (p - 1) (q - 1) 
    and not equal to p or q.
    """
    
    n = p * q # Just need to return n at the end
    found = 0
    e = 3 # start at 3 and stop when a good e is found

    while found == 0:
        gcd = Euclidean_Alg((p - 1)*(q - 1), e)
        if (gcd == 1) and (e != p) and (e != q):
            found = 1
        else:
            e += 1
    return(n, e)
````

````python
def Find_Private_Key_d(e, p, q):
    """
    Implement this method
    to find the decryption exponent d such that
    d is the modular inverse of e. 
    
    This will use the Extended Euclidean Algorithm
    
    This function should return the following:
    d: the decryption component.
    """
    
    # Extended Euclidean Algorithm:

    n = (p - 1)*(q - 1)
    s1, t1 = 1, 0  # The first iteration, this will give us just m i.e. s1*m + t1*n = 1m + 0
    s2, t2 = 0, 1  # The first iteration, this will give us just n i.e. s2*m + t2*n = 0 + 1n
    while n > 0:  # We'll finish when the remainder of e/n is 0
        k = e % n  # remainder
        quot = e // n  # quotient
        e = n
        n = k
        s1_hat, t1_hat = s2, t2  # We still need the old values of s1 and t1, so we use temp placeholder variables
        s2_hat, t2_hat = s1 - quot * s2, t1 - quot * t2
        s1, t1 = s1_hat, t1_hat
        s2, t2 = s2_hat, t2_hat
    # e is the GCD, s1 is the Bezout coefficient for e, aka our d for the private key
    # the GCD should be 1

    if e == 1:
        # Loop until we get to a positive modular inverse
        while s1 < 0:
            s1 += (p - 1)*(q - 1)
        return s1
    else:
        print("Something went wrong")
        return 0
````

### 3.
#### Putting things all together.

1. In this part, you will define two functions `Encode` and `Decode` which will use the public and private keys that you calculated using the above 2 functions in the second toolset.
2. Using the public key, the `Encode` function will encode a message and generate the corresponding cipher_text.
3. Using the private key, the `Decode` function will decode a ciper_text and recover the original message.



````python
def Encode(n, e, message):
    """
    Here, the message will be a string of characters.
    Use the function Convert_Text from 
    the basic tool set and get a list of numbers.
    
    Encode each of those numbers using n and e and
    return the encoded cipher_text.
    """
    
    # message is a string, so we want to convert to a list of integers
    uncoded_message_numbers = Convert_Text(message) 
    cipher_text = []
    for num in uncoded_message_numbers:
        cipher_text.append(FME(num, e, n)) # FME finds num^e mod n to encode each converted letter
    return cipher_text
````

````python
def Decode(n, d, cipher_text):
    """
    Here, the cipher_text will be a list of integers.
    First, you will decrypt each of those integers using 
    n and d.
    Later, you will need to use the function Convert_Num, that converts the integers to a string, 
    from the basic toolset to recover the original message as a string. 
    
    """
    decoded_num = []
    for num in cipher_text:
        decoded_num.append(FME(num, d, n)) # FME finds num^d mod n to decode each number

    message = Convert_Num(decoded_num) # Convert decoded numbers back to letters
    return message
````

### 6. FME Exploration - (CODE INTERVIEW type question)



RSA needs to use a Fast Modular Exponentiation function because as the base and exponent get larger, the built-in mod function `%` can't process as quickly.  With larger and larger values, the built-in mod function will take more and more time than FME.

As an example, let's break this public key both ways:

```

Cipher: [143, 567, 567, 701, 280, 109, 277, 280, 417, 567, 220, 280, 679, 567, 55, 167, 237, 280, 292, 109, 626, 701, 55, 167, 237, 280, 109, 167, 679, 280, 207, 277, 220, 382, 382, 23]
```

Let's populate some variables, and then I'll create an `FME` method that uses the built-in mod instead of FME.  We'll also need a `factorize()` method so we can find `p` and `q` from `n` (to help us break the code):

````python
def factorize(n):
    # n is a number, return the smallest factor of n
    for i in range (2,n-1):
        if n % i == 0:
            return (i, n//i)
    return -1
````

Here is my psuedorandom "function" method, i.e. what calculates `x^2 + a mod n`:

````python
def pseudo_random(x, a, n):
    # return x^2 + a mod n
    return (x*x + a) % n
````

Here is my Pollard's Rho method:

````python
from random import randint # we use Python's random package to generate our random "a"

def pollard(n):

    rando = randint(1, n) # The randomly generated "a" in our f(x) = x^2 + a mod n function
    a = 2
    b = pseudo_random(a, rando, n)
    while a != b:
        # a runs once
        a = pseudo_random(a, rando, n)

        # b runs twice as fast.
        b = pseudo_random(pseudo_random(b, rando, n), rando, n)

        p = Euclidean_Alg(abs(b - a), n)
        if p > 1:
            q = n // p
            return (p, q)
    return (-1, -1)
````
